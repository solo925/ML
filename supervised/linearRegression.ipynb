data pre-processing


```python
import numpy as numpy
import matplotlib.pyplot as mtp
import pandas as pd
```

```python
data_set = pd.read_csv('Salary_Data.csv')
```

```python
data_set
```

we need to extaract the dependent and independent variables from the given dataset.The independent variable is years of experience and the dependent variable is salary

```python
x = data_set.iloc[:,:-1].values #:-all rows :-all columns -1 exlude the last column from the dataset
y = data_set.iloc[:,:-1].values
x
y
```

splitting the dataset into training and test dataset

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size =1/3,random_state = 0)
```

Training Dataset

we are not performing fearture scaling
fitting the simple Linear Regression to the training dataset

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train,y_train)#we fit our regressor object to the training dataset so that the model can easily learn 
#the correlations between the predictor and the training dataset
```

Prediction of the test set

```python
y_pred=regressor.predict(x_test)
x_pred=regressor.predict(x_train)
```

Visualization THE training set results

```python
mtp.scatter(x_train,y_train,color="green")
mtp.plot(x_train,x_pred,color="red")
mtp.title("Salary vs Experience(Training Dataset)")
mtp.xlabel("Years of exoerience")
mtp.ylabel("Salary(in Ksh)")
mtp.show()
```

Vusualizing the Test set results

```python
mtp.scatter(x_test,y_test,color="green")
mtp.plot(x_train,x_pred,color="red")
mtp.title("Salary vs Experience(Training Dataset)")
mtp.xlabel("Years of exoerience")
mtp.ylabel("Salary(in Ksh)")
mtp.show()
```
